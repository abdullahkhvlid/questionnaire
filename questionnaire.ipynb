```python
def ask_question(question_text, options, correct_option):
    attempts = 2
    while attempts > 0:
        print("\n" + question_text)
        for idx, option in enumerate(options, 1):
            print(f"{idx}. {option}")
        answer = input(f"Select your answer (1-{len(options)}): ").strip()

        if not answer.isdigit():
            attempts -= 1
            if attempts > 0:
                print(f"Invalid input. Enter a number. {attempts} try left.")
            else:
                print("Invalid input. No tries left.")
            continue

        choice = int(answer)
        if choice == len(options):
            return False, True
        elif choice == correct_option:
            return True, False
        else:
            attempts -= 1
            if attempts > 0:
                print(f"Incorrect. {attempts} try left.")
            else:
                print("Incorrect. No tries left.")
    return False, False


def start_quiz():
    total_winnings = 0
    prize_per_question = 1000

    print("===== Welcome to the Ultimate Quiz Battle =====")

    questions = [
        {
            "text": "What is the chemical formula of water?",
            "options": ["H2O", "CH3", "H34A", "OK0", "Quit"],
            "correct": 1
        },
        {
            "text": "What is the symbol of oxygen?",
            "options": ["O2", "H", "NO", "O0", "Quit"],
            "correct": 1
        },
        {
            "text": "When was Pakistan founded?",
            "options": ["1947", "1999", "1623", "500BC", "Quit"],
            "correct": 1
        },
        {
            "text": "What is the capital of Australia?",
            "options": ["Sydney", "Melbourne", "Canberra", "Perth", "Quit"],
            "correct": 3
        },
        {
            "text": "How many continents are there?",
            "options": ["5", "6", "7", "8", "Quit"],
            "correct": 3
        }
    ]

    for q in questions:
        correct, quit_game = ask_question(q["text"], q["options"], q["correct"])
        if quit_game:
            print(f"\nYou chose to quit. Total winnings: Rs.{total_winnings}")
            break
        elif correct:
            total_winnings += prize_per_question
            print(f"‚úÖ Correct! Total winnings: Rs.{total_winnings}")
        else:
            print(f"‚ùå Game over. Final winnings: Rs.{total_winnings}")
            break
    else:
        print(f"\nüéâ You completed the Ultimate Quiz Battle! Final winnings: Rs.{total_winnings}")


start_quiz()
```
